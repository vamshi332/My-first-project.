```python
import pandas as pd
import io

# Data for Project DataFrame
project_data = """ID,Project,Cost,Status
A001,Project 1,1002000,Finished
A002,Project 2,2000000,Ongoing
A003,Project 3,4500000,Finished
A004,Project 4,5500000,Ongoing
A005,Project 5,,Finished
A002,Project 6,680000,Failed
A005,Project 7,400000,Finished
A003,Project 8,350000,Failed
A001,Project 9,,Ongoing
A003,Project 10,300000,Finished
A001,Project 11,2000000,Failed
A004,Project 12,1000000,Ongoing
A004,Project 13,3000000,Finished
A005,Project 14,200000,Finished
"""

# Data for Employee DataFrame
employee_data = """ID,Name,Gender,City,Age
A001,John Alter,M,Paris,25
A002,Alice Luxumberg,F,London,27
A003,Tom Sabestine,M,Berlin,29
A004,Nina Adgra,F,Newyork,31
A005,Amy Johny,F,Madrid,30
"""

# Data for Seniority Level DataFrame
seniority_data = """ID,Designation Level
A001,2
A002,2
A003,3
A004,2
A005,3
"""

# Create DataFrames from the data
project_df = pd.read_csv(io.StringIO(project_data))
employee_df = pd.read_csv(io.StringIO(employee_data))
seniority_df = pd.read_csv(io.StringIO(seniority_data))

# Save DataFrames to CSV files
project_df.to_csv('project_data.csv', index=False)
employee_df.to_csv('employee_data.csv', index=False)
seniority_df.to_csv('seniority_data.csv', index=False)

print("CSV files created successfully.")
```

```python
import pandas as pd

# Load the project DataFrame from the CSV file (assuming you saved it as 'project_data.csv')
project_df = pd.read_csv('project_data.csv')

# Calculate the running average for the 'Cost' column

# Initialize variables
running_sum = 0
count = 0

# Iterate through the 'Cost' column
for index, cost in enumerate(project_df['Cost']):
    if pd.isna(cost):  # Check if the value is NaN (missing)
        if count == 0:  # If no previous values, replace with 0 or handle as needed
            project_df.loc[index, 'Cost'] = 0  # Or choose another appropriate handling
        else:
            project_df.loc[index, 'Cost'] = running_sum / count  # Replace with running average
    else:
        running_sum += cost  # Add the valid cost to the running sum
        count += 1          # Increment the count of valid values

# Print the DataFrame with replaced missing values
print(project_df)

# Save the updated DataFrame to a new CSV file (optional)
project_df.to_csv('project_data_filled.csv', index=False)
```

```python
import pandas as pd

# Load the Employee DataFrame from the CSV file
employee_df = pd.read_csv('employee_data.csv')

# Split the "Name" column into "First Name" and "Last Name"
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', expand=True)

# Remove the original "Name" column
employee_df.drop('Name', axis=1, inplace=True)

# Print the modified DataFrame
print(employee_df)

# Save the modified DataFrame to a new CSV file (optional)
employee_df.to_csv('employee_data_split.csv', index=False)
```

```python
import pandas as pd

# Load the DataFrames from the CSV files
project_df = pd.read_csv('project_data.csv')
employee_df = pd.read_csv('employee_data.csv')
seniority_df = pd.read_csv('seniority_data.csv')

# Merge the project and employee DataFrames on the "ID" column
# We use 'left' join to include all projects, even if there's no matching employee
merged_df = pd.merge(project_df, employee_df, on='ID', how='left')

# Merge the merged DataFrame with the seniority DataFrame on the "ID" column
# Again, using 'left' join to keep all projects
final_df = pd.merge(merged_df, seniority_df, on='ID', how='left')

# Print the final merged DataFrame
print(final_df)

# Save the final DataFrame to a new CSV file (optional)
final_df.to_csv('final_data.csv', index=False)
```

```python
import pandas as pd

# Load the final DataFrame from the CSV file
final_df = pd.read_csv('final_data.csv')

# Calculate the bonus for employees who finished projects
# Create a new column 'Bonus' initialized with 0
final_df['Bonus'] = 0

# Identify the indices where 'Status' is 'Finished'
finished_project_indices = final_df[final_df['Status'] == 'Finished'].index

# Calculate the bonus (5% of 'Cost') for those indices
final_df.loc[finished_project_indices, 'Bonus'] = final_df.loc[finished_project_indices, 'Cost'] * 0.05

# Print the DataFrame with the new 'Bonus' column
print(final_df)

# Save the updated DataFrame to a new CSV file (optional)
final_df.to_csv('final_data_with_bonus.csv', index=False)
```

```python
import pandas as pd

# Load the final DataFrame from the CSV file
final_df = pd.read_csv('final_data.csv')

# Demote designation level for employees with failed projects
# Create a boolean mask for rows where the project status is 'Failed'
failed_projects_mask = final_df['Status'] == 'Failed'

# Use the mask to select rows and demote the 'Designation Level' by 1
# Ensure that the level does not go below 1
final_df.loc[failed_projects_mask, 'Designation Level'] = final_df.loc[failed_projects_mask, 'Designation Level'].apply(lambda x: max(1, x - 1))

# Delete employees with designation level above 4
# Create a boolean mask for rows where 'Designation Level' is greater than 4
above_level_4_mask = final_df['Designation Level'] > 4

# Use the mask to get the indices of rows to be deleted
rows_to_delete = final_df[above_level_4_mask].index

# Delete the rows using the indices
final_df.drop(rows_to_delete, inplace=True)

# Print the modified DataFrame
print(final_df)

# Save the modified DataFrame to a new CSV file (optional)
final_df.to_csv('final_data_modified.csv', index=False)
```

```python
import pandas as pd

# Load the final DataFrame from the CSV file
final_df = pd.read_csv('final_data.csv')

# Print column names to verify what's available
print("Available columns:", final_df.columns.tolist())

# Function to add "Mr." or "Mrs." based on gender
def add_title(row):
    if row['Gender'] == 'M':
        return 'Mr. ' + row['Name']  # Adjust column name here
    elif row['Gender'] == 'F':
        return 'Mrs. ' + row['Name']  # Adjust column name here
    else:
        return row['Name']  # Adjust column name here

# First, verify if the required columns exist
required_columns = ['Gender', 'Name']  # Adjust column names as needed
missing_columns = [col for col in required_columns if col not in final_df.columns]

if missing_columns:
    print(f"Error: Missing columns: {missing_columns}")
else:
    # Apply the function to create a new "Name" column with titles
    final_df['Name'] = final_df.apply(add_title, axis=1)
    
    # Drop the original "Gender" column
    final_df.drop('Gender', axis=1, inplace=True)
    
    # Print the modified DataFrame
    print(final_df)
    
    # Save the modified DataFrame to a new CSV file
    final_df.to_csv('final_data_titled.csv', index=False)
```

```python
import pandas as pd

# Load the final DataFrame from the CSV file
final_df = pd.read_csv('final_data.csv')

# Promote designation level using IF condition
for index, age in enumerate(final_df['Age']):
    if age > 29:
        # Promote the designation level by 1
        final_df.loc[index, 'Designation Level'] += 1

# Print the updated DataFrame
print(final_df)

# Save the updated DataFrame to a new CSV file (optional)
final_df.to_csv('final_data_promoted.csv', index=False)
```

```python
import pandas as pd

# Load the merged DataFrame from the CSV file (assuming you saved it as 'final_data.csv')
final_df = pd.read_csv('final_data.csv')

# Group by 'ID' and 'Name' and calculate the sum of 'Cost'
total_cost_df = final_df.groupby(['ID', 'Name'])['Cost'].sum().reset_index()

# Rename the 'Cost' column to 'Total Cost'
total_cost_df.rename(columns={'Cost': 'Total Cost'}, inplace=True)

# Print the resulting DataFrame
print(total_cost_df)

# Save the resulting DataFrame to a new CSV file (optional)
total_cost_df.to_csv('TotalProjCost.csv', index=False)
```
